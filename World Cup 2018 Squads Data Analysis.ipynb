```python
#Import library
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read dataset and show some datas
os.chdir("D:\Work and School\Project\Datasets")
df= pd.read_excel("world_cup_2018_squads.xlsx")
```

```python
#Drop the "type" column
df = df.drop(['Type'], axis=1)

#Drop duplicate data
df= df.drop_duplicates()
```

```python
#Show data to check the new data
df
```

```python
#Create new columns from splitting the "DOB" column
df[['Year of Birth','Month of Birth', 'Date of Birth']] = df.DOB.str.split("-", expand=True)
```

```python
#Show data to check the new columns
df
```

```python
# Change the column type to numeric
df["Year of Birth"] = pd.to_numeric(df["Year of Birth"])
df["Month of Birth"] = pd.to_numeric(df["Month of Birth"])
df["Date of Birth"] = pd.to_numeric(df["Date of Birth"])
```

```python
# Create "Age" column
df['Age'] = 2018 - df['Year of Birth']
```

```python
#Show data to check the new column
df
```

```python
#Rename column name
df.rename(columns={'Country and Club':'CountryClub'}, inplace=True)
```

```python
#Create new columns from splitting the "CountryClub" column
df[['Country','Club']] = df.CountryClub.str.split(" ", n= 1, expand=True)
```

```python
#Show Data to check the new column
df
```

```python
#Change column order
df = df[['Team', "Group", "Position", "Name", "DOB", "Year of Birth", "Month of Birth", "Date of Birth", "Age", "Caps", "Goals", "CountryClub", "Country", "Club"]]
```

```python
#Show data after changing column order
df
```

```python
#Number of Teams and Groups
TeamGroup= df[df.columns[0:2]].drop_duplicates(subset=['Team', 'Group'])
sumOfTeam= TeamGroup['Team'].count()
groupNoDuplicate= TeamGroup.drop_duplicates(subset=['Group'], keep='last')
groupOnly= groupNoDuplicate.drop(['Team'], axis=1)
sumOfGroup= groupOnly['Group'].count()

print("Number of Teams: ", sumOfTeam)
print("Number of Groups: ", sumOfGroup)
```

```python
#Detailed average age, caps, and goals
avgage= df["Age"].mean()
avgcaps= df["Caps"].mean()
avggoals= df["Goals"].mean()

print("Average Age     : ", avgage)
print("Average Caps    : ", avgcaps)
print("Average Goals   : ", avggoals)
```

```python
#Check players who are above average age
df[df.Age >= avgage]
```

```python
#A histogram of the number of players by age who play in the World Cup 2018
sns.histplot(df['Age'], kde=False, color='red', bins=100)
plt.title('Number of Players by Age', fontsize=14)
plt.xlabel('Age', fontsize=10)
plt.ylabel('Number of Players', fontsize=10)
```

```python
#Check players who are above average caps
df[df.Caps >= avgcaps]
```

```python
#A histogram of the number of players by caps who play in the World Cup 2018
sns.histplot(df['Caps'], kde=False, color='red', bins=100)
plt.title('Number of Players by Caps', fontsize=14)
plt.xlabel('Number of Caps', fontsize=10)
plt.ylabel('Number of Players', fontsize=10)
```

```python
#Check players who are above average goals
df[df.Goals >= avggoals]
```

```python
#A histogram of the number of players by goals who play in the World Cup 2018
sns.histplot(df['Goals'], kde=False, color='red', bins=100)
plt.title('Number of Players by Goals', fontsize=14)
plt.xlabel('Goals', fontsize=10)
plt.ylabel('Number of Players', fontsize=10)
```

```python
#France players participated in the World Cup 2018
francePlayers= df[df.Team.isin(["France"])]
display(francePlayers)
```

```python
#France Team Average Age
sumFranceAge= francePlayers['Age'].sum()
countFrancePlayers= francePlayers['Name'].count()
franceAvgAge= sumFranceAge/countFrancePlayers
print("France Average Age: ", franceAvgAge)
```

```python
#France Team Average Caps
sumFranceCaps= francePlayers['Caps'].sum()
countFrancePlayers= francePlayers['Name'].count()
franceAvgCaps= sumFranceCaps/countFrancePlayers
print("France Average Caps: ", franceAvgCaps)
```

```python
#France Team Average Goals
sumFranceGoals= francePlayers['Goals'].sum()
countFrancePlayers= francePlayers['Name'].count()
franceAvgGoals= sumFranceGoals/countFrancePlayers
print("France Average Goals: ", franceAvgGoals)
```

```python
#Croatia players participated in the World Cup 2018
croatiaPlayers= df[df.Team.isin(["Croatia"])]
display(croatiaPlayers)
```

```python
#Croatia Team Average Age
sumCroatiaAge= croatiaPlayers['Age'].sum()
countCroatiaPlayers= croatiaPlayers['Name'].count()
croatiaAvgAge= sumCroatiaAge/countCroatiaPlayers
print("Croatia Average Age: ", croatiaAvgAge)
```

```python
#Croatia Team Average Caps
sumCroatiaCaps= croatiaPlayers['Caps'].sum()
countCroatiaPlayers= croatiaPlayers['Name'].count()
croatiaAvgCaps= sumCroatiaCaps/countCroatiaPlayers
print("Croatia Average Caps: ", croatiaAvgCaps)
```

```python
#Croatia Team Average Goals
sumCroatiaGoals= croatiaPlayers['Goals'].sum()
countCroatiaPlayers= croatiaPlayers['Name'].count()
croatiaAvgGoals= sumCroatiaGoals/countCroatiaPlayers
print("Croatia Average Goals: ", croatiaAvgGoals)
```

```python
#France Average Age vs Croatia Average Age
Team = ['France','Croatia']
avgAge = [franceAvgAge, croatiaAvgAge]

New_Colors = ['blue','orange']
plt.bar(Team, avgAge, color=New_Colors)
plt.title('France Average Age vs Croatia Average Age', fontsize=14)
plt.xlabel('Team', fontsize=10)
plt.ylabel('Average Age', fontsize=10)
plt.show()
```

```python
#France Average Caps vs Croatia Average Caps
Team = ['France','Croatia']
avgCaps = [franceAvgCaps, croatiaAvgCaps]

New_Colors = ['blue','orange']
plt.bar(Team, avgCaps, color=New_Colors)
plt.title('France Average Caps vs Croatia Average Caps', fontsize=14)
plt.xlabel('Team', fontsize=10)
plt.ylabel('Average Caps', fontsize=10)
plt.show()
```

```python
#France Average Goals vs Croatia Average Goals
Team = ['France','Croatia']
avgGoals = [franceAvgGoals, croatiaAvgGoals]

New_Colors = ['blue','orange']
plt.bar(Team, avgGoals, color=New_Colors)
plt.title('France Average Goals vs Croatia Average Goals', fontsize=14)
plt.xlabel('Team', fontsize=10)
plt.ylabel('Average Goals', fontsize=10)
plt.show()
```
